```python
def get_booth_and_constituency(file_path):

    #extract filename from filepath
    if ".pdf" in file_path:
        filename = file_path.split("/")[-1]

        # Use regular expression to find numbers after "AC" and "PS" prefixes
        matches_ac = re.search(r'AC-(\d+)', filename)
        matches_ps = re.search(r'PS-(\d+)', filename)

        # Extract the required numbers if available
        number_ac = int(matches_ac.group(1)) if matches_ac else None
        number_ps = int(matches_ps.group(1)) if matches_ps else None
    
    return number_ac, number_ps
```

```python
def get_processed_excel(file_path):

    if ".xlsx" in file_path:
        voting_data = pd.read_excel(file_path, engine="openpyxl", header=None, index_col=0)

        voting_processed = voting_data.drop(['Total Voter', 'Interval', 'Random No.'])
        voting_processed = voting_processed.round().astype(int)

        voting_processed = voting_processed.transpose()

        return voting_processed
    
    else:

        raise ValueError("File is not in the .xlsx format, check the format once!")
```
